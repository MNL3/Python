```python
df = pd.read_csv('kc_house_data.csv', encoding='windows-1251', sep=',')
df.head()
```

### 1.1 
Создать новый признак price_per_sq_lot,
который будет содержать среднюю стоимость за
один кв. метр общей площади

```python
df['price_per_sq_lot'] = round((df['price']/df['sqft_lot']), 2)
df.head(5)
```

### 1.2 
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = 0
df.loc[df['yr_renovated'] != 0, 'delta_renovated'] = df['yr_renovated'] - df['yr_built']
```

```python
df.head(5)
```

### 1.3 
Создайте признаки года продажи, месяца продажи

```python
import calendar
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year_sale'] = df['date'].apply(lambda x: x.date().year)
df['month_sale'] = df['date'].apply(lambda x: x.date().month)
d = dict(enumerate(calendar.month_abbr))
df['month_sale'] = df['month_sale'].map(d)

df.head(5)
```

### 
1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
```

### 3.1 
Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price':'mean'}).sort_values('price')
```

### 3.2 
Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

### 3.3 
Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.head(3)
```

```python
df.pivot_table(index='waterfront', columns='view', aggfunc='count', values='price')
```

### 3.4 
Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms'])
```

### 3.5 
Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.groupby(['condition', 'grade']).agg({'price': 'median'}).unstack('condition').fillna(0)
```

```python
df.pivot_table(index='grade', columns='condition', values='price', aggfunc='median').fillna(0)
```
